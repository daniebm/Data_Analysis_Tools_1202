```python
#Importing Python Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
print("Done")
```

```python
#Loading Dataset
df_youtube = pd.read_csv('youtube_dataset.csv')
df_youtube.head()
```

```python
# Sorting Data Based on Suscribers
df_sorted = df_youtube.sort_values(by=['subscribers'],ascending = False)
df_sorted.head()
```

```python
#Subsetting the top 1000 Channels
df_filtered = df_sorted.iloc[:1000]
df_filtered.head()
```

```python
# Function to calculating the distirbution of All Channel Type
# Data Distribution
print(df_youtube.groupby('channeltype')['channeltype'].count().reset_index(name='Channel Count').sort_values(['Channel Count'],
     ascending=False))

# Bar Chart
df_youtube['channeltype'].value_counts().plot(kind='bar')
plt.ylabel('Channel Count')
plt.xlabel('Channel Type')
plt.title('Channel Type Distribution - All (Before)')
```

```python
# Function to calculating the distirbution of Top 1000 Channel Type
# Data Distribution
print(df_filtered.groupby('channeltype')['channeltype'].count().reset_index(name='Channel Count').sort_values(['Channel Count'],
      ascending=False))

# Bar Chart
df_filtered['channeltype'].value_counts().plot(kind='bar')
plt.ylabel('Channel Count')
plt.xlabel('Channel Type')
plt.title('Channel Type Distribution - Top 1000 (After)')
```

```python
#Loading Top 1000 Channels to CSV
df_filtered.to_csv('youtubechannel_top_1000.csv')
```
